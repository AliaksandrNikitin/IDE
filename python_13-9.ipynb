```python
# 9.8

import pandas as pd
import numpy as np

import plotly.express as px
import plotly.graph_objects as go
import plotly.figure_factory as ff

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
churn_data = pd.read_csv('data/churn.csv')
data = churn_data.copy()
display(data)


exited_by_active_status = data.groupby(
    'IsActiveMember')['Exited'].agg('mean')

fig = plt.figure(figsize=(5, 5))
axes = fig.add_axes([0, 0, 1, 1])
axes.pie(
    exited_by_active_status,
    labels=['Не активный статус', ' активный статус'],
    autopct='%.1f%%',
    explode = [0.1, 0],
);

axes.set_title(u'Cоотношение ушедших \nпо наличию статуса активного члена', fontsize=14)
exited_by_active_status
```

#9.9 доля ушедших клиентов больше в Германии. это может быть связанно в связи с тем что на Германском рынке бынк есть более привлекательные банки

```python
# 9.10

import pandas as pd
import numpy as np

import plotly.express as px
import plotly.graph_objects as go
import plotly.figure_factory as ff

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
churn_data = pd.read_csv('data/churn.csv')
data = churn_data.copy()



def get_credit_score_cat(credit_score):
    if credit_score >= 300 and credit_score < 500:
        return "Very_Poor"
    elif credit_score >= 500 and credit_score < 601:
        return "Poor"
    elif credit_score >= 601 and credit_score < 661:
        return "Fair"
    elif credit_score >= 661 and credit_score < 781:
        return "Good"
    elif credit_score >= 851:
        return "Top"
    elif credit_score >= 781 and credit_score < 851:
        return "Excellent"
    elif credit_score < 300:
        return "Deep"

data['CreditScoreCat'] = data['CreditScore'].apply(get_credit_score_cat)

display(data)

pivot_table = data.pivot_table(
    values='Exited',
    index='CreditScoreCat',
    columns='Tenure',
    aggfunc='mean'
).round(2)
fig = px.imshow(
    pivot_table,
    text_auto=True
    ).update_layout(
        title={'text':'Статистика оттока в зависимости от кредитного рейтинга и количества лет пользования услугами банка'},
        xaxis_title={'text': 'Количество лет использования услуг'},
        yaxis_title={'text': 'Кредитный рейтинг'}
        )
fig.show()
```
